```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("real_estate_price_size (1).csv")
```

```python
df.isnull().sum()
```

```python
x = df.drop("price", axis = 1)
y = df["price"]
```

```python
plt.scatter(x, y, facecolor = "black",color = "red")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```
